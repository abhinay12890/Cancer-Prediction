```python
o=['Age','Survival Months','Reginol Node Positive','Regional Node Examined','Tumor Size']
for x in o:
    irq=train[x].quantile(0.75)-train[x].quantile(0.25)
    up=train[x].quantile(0.75)+1.5*irq
    lp=train[x].quantile(0.25)-1.5*irq
    train=train[(train[x]<up) & (train[x]>lp)]
train
```

```python
y=train['Status']
x=train.drop('Status',axis="columns")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
train.drop(['Race','Marital Status','N Stage','6th Stage','differentiate','Tumor Size','Estrogen Status'],axis="columns",inplace=True)
```

```python
train
```

```python
y=train["Status"]
x=train.drop("Status",axis="columns")
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smote=SMOTE(sampling_strategy="minority")
xn,yn=smote.fit_resample(x,y)

yn.value_counts()
```

```python
xn
```

```python
from sklearn.preprocessing import StandardScaler
nor=StandardScaler()
for qw in xn.columns:
    a=np.array(xn[qw]).reshape(-1,1)
    nor.fit(a)
    xn[qw]=nor.transform(a).reshape(-1,1)
xn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xn,yn,random_state=42,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

```python
rpred=model.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

```python
tpred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print("E Random:",mean_squared_error(y_test,rpred)**0.5)
print("E Tree:",mean_squared_error(y_test,tpred)**0.5)
```
